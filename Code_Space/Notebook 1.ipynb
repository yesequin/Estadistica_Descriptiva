### Tipos de datos

Todos los datos existen en 2 categorías fundamentales:

- Categóricos: género, categoría de película, método de pago

- Numéricos: edad, altura, temperatura

- Categóricos:

### Medidas de tendencia central

Son una manera de resumir información.

Las medidas de tendencia central más importantes son:

- Media (promedio): es susceptible a valores atípicos. Por ejemplo, hablar de ingreso per capita no va a demostrar la realidad de un pais (metáfora de Bill Gates).
    
    ### Media = $$ \frac{1}{N} \sum_{i=1}^N x_i  $$ 
    ![Media](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%202.png)
    
    
    Siendo N el tamaño de la población

````python
cars['price_usd'].mean() 
#Promedio de la columna Price_usd
````

- Mediana: dato central. Recordemos que el conjunto debe estar ordenado (de menor a mayor):
    
    ### Mediana(impar) = $$ x_{(n+1)/2}^{\text{ordered}} $$
    ### Mediana(par) = $$ \frac{x_{n/2}^{\text{ordered}} + x_{n/2+1}^{\text{ordered}}}{2} $$
    ![Mediana](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%203.png)

````python
cars['price_usd'].median() 
#Mediana de la columna Price_usd
````

- Moda: dato que más se repite. La moda no aplica para datos numéricos continuos.
    
    ### Moda = $$ x_k $$ donde $$ \text{Freq}(x_k) = \max{(\text{Freq}(x_i))} $$ 
    ![Moda](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%204.png)

Usamos un **diagrama de frecuencias**, la cual usa los datos numéricos discretos o continuos para mostrar la frecuencia en que un dato se permite. Todos los estadísticos descriptivos se pueden dibujar o representar en términos de este diagrama.

El diagrama de frecuencias (gráfico de barras) sirve para representar las variables cualitativas ordinales. En el eje horizontal se representan las diferentes categorí­as y sobre él se levantan unas columnas o barras cuya altura es proporcional a la frecuencia de cada categorí­a. También podrí­amos utilizar este tipo de gráfico para representar variables cuantitativas discretas, pero lo que no es correcto hacer es usarlo para las variables cualitativas nominales. El gran mérito del diagrama de barras es expresar la magnitud de las diferencias entre las categorí­as de la variable. Pero ahí­ está precisamente, su punto débil, ya que son fácilmente manipulables si modificamos los ejes.
El **histograma** es un gráfico con un significado mucho más profundo. Un histograma representa una distribución de frecuencias que se utiliza (o debe) para representar la frecuencia de las variables cuantitativas continuas. Aquí­ no es la altura, sino el área de la barra lo que es proporcional a la frecuencia de ese intervalo, y está en relación con la probabilidad con la que cada intervalo puede presentarse.
Otra diferencia muy importante entre el diagrama de barras y el histograma es que en el primero solo se representan los valores de las variables que hemos observado al hacer el estudio. Sin embargo, el histograma va mucho más allá, ya que representa todos los valores posibles que existen dentro de los intervalos, aunque no hayamos observado ninguno de forma directa. Permite así­ calcular la probabilidad de que se represente cualquier valor de la distribución, lo que es de gran importancia si queremos hacer inferencia y estimar valores de la población a partir de los resultados de nuestra muestra.

La estadística descriptiva no son solo números, son visualizaciones. Esto nos dice que la estadística descriptiva se divide en dos ramas fundamentales:

- Números concretos con los que se resume información
- Visualizaciones que nos permiten describir de forma visual y resumida la información contenida en un conjunto de datos.

### Medidas de dispersión

Las 3 medidas de dispersión más famosas y más utilizadas son:

- **Rango**: La manera más rápida de medir la dispersión de un conjunto de datos es cogiendo el valor mínimo y máximo y hacer una resta de ellos; al hacer la resta ya tengo la máxima medida de la dispersión de ese conjunto: esa medida es lo que llamamos el rango.

````python
#Rango = Max - Min
rango = cars['price_usd'].max()-cars['price_usd'].min()
rango
````

- **Rango intercuartil**: se define en base en algo llamado cuartil. Un cuartil tiene como propósito definir el conjunto de datos en términos de 4 subdivisiones homogeneas (misma cantidad de datos en cada subdivisión)
    - Q1: la mitad entre la mediana y el valor mínimo.
    - Q2: la mediana.
    - Q3: la mitad entre la mediana y el valor máximo.
    - IQR: Rango intercuartílico. La distancia entre Q3 y Q1 es lo que se llama rango intercuartil.

![Untitled](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%205.png)

El diagrama de caja es la visualización por excelencia para tener una representación simplificada de la dispersión de los datos en referencia a la mediana en particular.

````python
#Quartiles
median=cars['price_usd'].median() #recordemos que la mediana  Q2
Q1=cars['price_usd'].quantile(q=0.25)
Q3=cars['price_usd'].quantile(q=0.75)
min_val=cars['price_usd'].quantile(q=0)
max_val=cars['price_usd'].quantile(q=1.0)
print('Valor mínimo:',min_val)
print('Cuartil 1:',Q1)
print('Media:',median)
print('Cuartil 3:',Q3)
print('valor máximo:',max_val)
````

````python
#Rango intercuartílico (IQR):
iqr= Q3-Q1
iqr
````

- **Desviación estándar**: medida más famosa y más usada en el ámbito de la estadística descriptiva. También llamado Sigma. Es otra medida para calcular la dispersión de los datos de un conjunto cualquiera.
    
    ![Untitled](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%206.png)

````python
cars['price_usd'].std()
````

Tenemos varios tipos de distribuciones que cambia el cálculo de la desviación estándar:

- Distribución normal: Tiene la forma de una campana (campana de Gauss), y es perfectamente simétrica. Para el caso de una distribución normal el valor de la mitad es la mediana o Q2. También, la media y la mediana van a coincidir.
    
    ![Untitled](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%207.png)
    
    El sigma también mide desviaciones. La teoría de la distribución normal nos dice que cuando consideramos todos los datos contenidos entre el promedio menos 3 veces la desviación estándar y el promedio mas 3 veces la desviación estándar ahí estamos contemplando la totalidad de los datos (99,72%), y lo demás serían los outliers. Los outliers están demasiado lejos (3 veces sigma de lejos).
    
    Esto se mapea en términos del rango intercuartil con el siguiente método (**método de detección de outliers con el rango intercuartil**):
    
    ![Untitled](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%208.png)
    
    
    
    **El max es con Q3**
    
    Esto recorta las barras y todo lo que esté por fuera de las barras se puede identificar como datos anómalos. El 1.5 es el valor que, cuando lo correspondemos con los límites anteriores (µ-3σ y µ+3σ) coinciden (son muy cercanos).
    


````python
minlimit= Q1 - 1.5*iqr
maxlimit= Q3 + 1.5*iqr
print(f"rango para detección de outliers: {minlimit}, {maxlimit}")
````

- Distribución no simétrica: un brazo de la caja debe ser más grande que otro. Se generaliza la noción de cómo definir el criterio a partir del cual los puntos se consideran anómalos.
    
    ![Untitled](/workspaces/Estadistica_Descriptiva/images_estadistica_descriptiva/Untitled%209.png)

Ahora podremos dibujar el diagrama de caja de la distribución en nuestro cars DataFrame, y tener claro si estamos ante unos datos con distribución sesgada o no sesgada

````python
import timeit #sirve para medir el performance de los modelos
from sklearn import datasets #nos permite descargar un dataset
from sklearn import linear_model #nos permite hacer una regresión linear sencilla solo para validar el efecto de transformar los datos.

X, y = datasets.load_diabetes(return_X_y=True) #este es el dataset
raw = X[:, None, 2] #le agregamos una transformación en las dimensiones para que se ajuste al formato de entrada al momento del pre procesamiento
#detalles del Dataset: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html
````

## Reglas de escalamiento

- max-min scaling: mejor para datos uniformemente distribuidos


- z-score scaling: mejor para datos distribuidos "normalmente" (forma de campana de gauss)

Ese es el Estandar, pero hay que validar estas afirmaciones para nuestros datos en específico.

````python
# REGLA DEL MÁX - MÍN:
##raw son los datos originales (x)
max_raw = max(raw)
min_raw = min(raw)
scaled = (((2*raw)-min_raw-max_raw)/(max_raw-min_raw))

##Gráfico de datos originales VS datos escalados:
fig, axs = plt.subplots(2,1, sharex= True)
axs[0].hist(raw)
axs[1].hist(scaled)

##Modelos para entrenamiento:
def train_raw():
    linear_model.LinearRegression().fit(raw, y)

def train_scaled():
    linear_model.LinearRegression().fit(scaled, y)

##Calculando los tiempos que se demora cada modelo:
raw_time = timeit.timeit(train_raw, number = 100)
scaled_time = timeit.timeit(train_raw, number = 100)
print('trainning time for raw data : {} '.format(raw_time))
print('trainning time for scaled data : {}'.format(scaled_time))
````

````python
#REGLA DEL Z-SCORE:

std_dev = np.std(raw) #desviación estándar
prom = np.mean(raw) #media o promedio

zScaled= (raw - prom)/std_dev

##Gráfico de datos originales VS datos escalados:
fig, axs = plt.subplots(2,1, sharex= True)
axs[0].hist(raw)
axs[1].hist(zScaled)

##Modelos para entrenamiento:
def train_raw():
    linear_model.LinearRegression().fit(raw, y)

def train_z_scaled():
    linear_model.LinearRegression().fit(z_scaled, y)

##Calculando los tiempos que se demora cada modelo:
raw_time = timeit.timeit(train_raw, number = 100)
z_scaled_time = timeit.timeit(train_raw, number = 100)
print('trainning time for raw data : {} '.format(raw_time))
print('trainning time for z_scaled data : {}'.format(z_scaled_time))
````
