### Tipos de datos

Todos los datos existen en 2 categorías fundamentales:

- Categóricos: género, categoría de película, método de pago
    - Ordinal: cuando exise una relación de orden entre las categorías
    - Nominal
- Numéricos: edad, altura, temperatura
    - Discretos: tipos de variables enteras (int)
    - continuos: tipos de variables flotantes (float)

Un notebook es un ambiente ideal para explorar la información.

Los tipos de datos se identifican de la siguiente manera:

- Categóricos:
    - `object`: es una variable textual que NO son números
    - `bool`: True y False
- Numéricos:
    - `int64`: Numérico discreto. Formato por defecto de pandas para las variables numéricas enteras.
    - `float64`: Formato por defecto de pandas para las variables continuas.

Usamos un **diagrama de frecuencias**, la cual usa los datos numéricos discretos o continuos para mostrar la frecuencia en que un dato se permite. Todos los estadísticos descriptivos se pueden dibujar o representar en términos de este diagrama.

El diagrama de frecuencias (gráfico de barras) sirve para representar las variables cualitativas ordinales. En el eje horizontal se representan las diferentes categorí­as y sobre él se levantan unas columnas o barras cuya altura es proporcional a la frecuencia de cada categorí­a. También podrí­amos utilizar este tipo de gráfico para representar variables cuantitativas discretas, pero lo que no es correcto hacer es usarlo para las variables cualitativas nominales. El gran mérito del diagrama de barras es expresar la magnitud de las diferencias entre las categorí­as de la variable. Pero ahí­ está precisamente, su punto débil, ya que son fácilmente manipulables si modificamos los ejes.
El **histograma** es un gráfico con un significado mucho más profundo. Un histograma representa una distribución de frecuencias que se utiliza (o debe) para representar la frecuencia de las variables cuantitativas continuas. Aquí­ no es la altura, sino el área de la barra lo que es proporcional a la frecuencia de ese intervalo, y está en relación con la probabilidad con la que cada intervalo puede presentarse.
Otra diferencia muy importante entre el diagrama de barras y el histograma es que en el primero solo se representan los valores de las variables que hemos observado al hacer el estudio. Sin embargo, el histograma va mucho más allá, ya que representa todos los valores posibles que existen dentro de los intervalos, aunque no hayamos observado ninguno de forma directa. Permite así­ calcular la probabilidad de que se represente cualquier valor de la distribución, lo que es de gran importancia si queremos hacer inferencia y estimar valores de la población a partir de los resultados de nuestra muestra.

La estadística descriptiva no son solo números, son visualizaciones. Esto nos dice que la estadística descriptiva se divide en dos ramas fundamentales:

- Números concretos con los que se resume información
- Visualizaciones que nos permiten describir de forma visual y resumida la información contenida en un conjunto de datos.

```python
import timeit #sirve para medir el performance de los modelos
from sklearn import datasets #nos permite descargar un dataset
from sklearn import linear_model #nos permite hacer una regresión linear sencilla solo para validar el efecto de transformar los datos.

X, y = datasets.load_diabetes(return_X_y=True) #este es el dataset
raw = X[:, None, 2] #le agregamos una transformación en las dimensiones para que se ajuste al formato de entrada al momento del pre procesamiento
#detalles del Dataset: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html
```

## Reglas de escalamiento

- max-min scaling: mejor para datos uniformemente distribuidos


- z-score scaling: mejor para datos distribuidos "normalmente" (forma de campana de gauss)

Ese es el Estandar, pero hay que validar estas afirmaciones para nuestros datos en específico.

```python
# REGLA DEL MÁX - MÍN:
##raw son los datos originales (x)
max_raw = max(raw)
min_raw = min(raw)
scaled = (((2*raw)-min_raw-max_raw)/(max_raw-min_raw))

##Gráfico de datos originales VS datos escalados:
fig, axs = plt.subplots(2,1, sharex= True)
axs[0].hist(raw)
axs[1].hist(scaled)

##Modelos para entrenamiento:
def train_raw():
    linear_model.LinearRegression().fit(raw, y)

def train_scaled():
    linear_model.LinearRegression().fit(scaled, y)

##Calculando los tiempos que se demora cada modelo:
raw_time = timeit.timeit(train_raw, number = 100)
scaled_time = timeit.timeit(train_raw, number = 100)
print('trainning time for raw data : {} '.format(raw_time))
print('trainning time for scaled data : {}'.format(scaled_time))
```

```python
#REGLA DEL Z-SCORE:

std_dev = np.std(raw) #desviación estándar
prom = np.mean(raw) #media o promedio

zScaled= (raw - prom)/std_dev

##Gráfico de datos originales VS datos escalados:
fig, axs = plt.subplots(2,1, sharex= True)
axs[0].hist(raw)
axs[1].hist(zScaled)

##Modelos para entrenamiento:
def train_raw():
    linear_model.LinearRegression().fit(raw, y)

def train_z_scaled():
    linear_model.LinearRegression().fit(z_scaled, y)

##Calculando los tiempos que se demora cada modelo:
raw_time = timeit.timeit(train_raw, number = 100)
z_scaled_time = timeit.timeit(train_raw, number = 100)
print('trainning time for raw data : {} '.format(raw_time))
print('trainning time for z_scaled data : {}'.format(z_scaled_time))
```
